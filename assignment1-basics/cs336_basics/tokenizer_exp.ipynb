```python
import torch
import pickle as pkl
from pathlib import Path
import json
import base64
import random
```

```python
path1 = "../data/owt_train_32000_merges.pkl"
path2 = "../data/owt_train_32000_vocab.json"
path3 = "../data/tiny_stories_10000_merges.pkl"
path4 = "../data/tiny_stories_10000_vocab.json"

vocab_str = Path(path2).read_text()
raw_tok2id = json.loads(vocab_str)
vocab: dict[int, bytes] = {
    int(idx): base64.b64decode(token.encode("ascii")) for idx, token in raw_tok2id.items()
}
list(vocab.items())[:10]
```

```python
from cs336_basics.tokenizer import Tokenizer
end_of_text_token = "<|endoftext|>"
tokenizer = Tokenizer.from_files(path2, path1, [end_of_text_token])
tokenizer2 = Tokenizer.from_files(path4, path3, [end_of_text_token])
```

```python
data_path1 = "../data/owt_valid.txt"
data_path2 = "../data/TinyStoriesV2-GPT4-valid.txt"

tot = []
tot2 = []
with open(data_path1, "r") as f:
    for i in range(10):
        f.seek(random.randint(0,10000))
        zz = f.read(1000)
        # 1000 bytes
        # print(zz[:100])
        encoded = tokenizer.encode(zz)
        print(len(encoded), encoded)
        tot.append(len(encoded))

        encoded = tokenizer2.encode(zz)
        print(len(encoded), encoded)
        tot2.append(len(encoded))

        # print(len(bytes(zz, encoding="latin-1")), zz)
print(1000* 10 / sum(tot))
print(1000* 10 / sum(tot2))
```
